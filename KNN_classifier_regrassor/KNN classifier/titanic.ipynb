## import libraries

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
```

## read Titanic Dataset

```python
df=sns.load_dataset("titanic")
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

## Drop the column on deck

```python
df.drop("deck",axis=1,inplace=True)
```

```python
df.isnull().sum()
```

## fill the value in age column

```python
df[['age']] = df[['age']].mean()
```

## fill the value in embark_town column

```python
my_list = 'Southampton'
df['embark_town'] = df['embark_town'].fillna(my_list)
```

## fill the value in age embarked column

```python
my_list = ['S','C','Q']
for i in my_list:
    df['embarked'] = df['embarked'].fillna(i)
```

```python
df
```

```python
df.drop("alive",axis=1,inplace=True)
```

```python
df
```

```python
df.columns
```

## using LabelEncoder convert string value to int

```python
l=['sex','embarked','class','who','adult_male','embark_town','alone']
for i in l:
    from sklearn.preprocessing import LabelEncoder
    le=LabelEncoder()
    le.fit(df[i])
    df[i]=le.transform(df[i])
```

```python
df
```

## seprate a dependant and independant variable (x=independant,y=dependant)

```python
x=df.iloc[:,1:]
```

```python
x
```

```python
y=df.iloc[:,0]
```

```python
y
```

## train_test_split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4)
```

## Splitting and Fitting The Data

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn
```

## fit the model

```python
knn.fit(x_train,y_train)
```

## Model Prediction

```python
Rs=knn.predict(x_test)
```

```python
Rs
```

## accuracy_score

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,Rs)
```

## confusion_matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,Rs)
```

## classification_report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, Rs))
```

## using GridSearchCV find best parameter

```python
from sklearn.model_selection import GridSearchCV
parameters = {'n_neighbors':np.arange(1,30,2)}
mx = GridSearchCV(knn, parameters)
print(mx.fit(x,y))
```

```python
mx.best_params_
```

```python
mx.best_score_
```
