## import used libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Read Dataset

```python
df=pd.read_csv("bank-full.csv")
```

```python
df
```

```python
# find a null values
df.isna().sum()
```

```python
#describe data in detrail
df.describe()
```

```python
# using label encoding convert string formated data into numeric
l=["job","marital","education","default","housing","loan","contact","month","poutcome","Target"]
for i in l:
    from sklearn.preprocessing import  LabelEncoder
    le= LabelEncoder()
    df[i]=le.fit_transform(df[i])
    df[i]
```

```python
df
```

## separate a X (Independant) and y (Dependant) variable

```python
x=df.drop("Target",axis=1)
```

```python
y=df["Target"]
```

```python
y.head(4)
```

## Split the data in train and test side for prediction

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=21)
```

```python
x_train.head()
```

## fitting the logestic regressor model

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression(random_state=0)
```

## fit the model on trainig data

```python
LR.fit(x_train,y_train)
```

## predict the output using test dataset

```python
y_pred=LR.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
## check accuracy score and confusion metrics
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print(accuracy_score(y_test,y_pred))
confusion_matrix(y_test,y_pred)
```
